```python
import requests
import json

# Replace YOUR_VIDEO_ID with the ID of the video you want to extract comments from
# https://www.youtube.com/watch?v=4b6yYQqL8zs
video_id = "4b6yYQqL8zs" 

# Set up the YouTube API request
api_key = "AIzaSyAFGXgs05BRR4QwJjTIgf8-BiiRm1SMc-0"
api_url = "https://www.googleapis.com/youtube/v3/commentThreads"

all_comments = []

# Set up the pagination parameters
params = {
    "part": "snippet",
    "videoId": video_id,
    "key": api_key,
    "maxResults": 100,  # The maximum number of comments that can be retrieved per API request
}

# Make the initial API request
response = requests.get(api_url, params=params)

# Extract the comments from the response
comments = response.json()["items"]
all_comments += comments

# Check if there are more comments to retrieve
next_page_token = response.json().get("nextPageToken")

# Keep making API requests until all comments have been retrieved
while next_page_token:
    # Update the pagination parameters
    params["pageToken"] = next_page_token

    # Make the next API request
    response = requests.get(api_url, params=params)

    # Extract the comments from the response
    comments = response.json()["items"]
    all_comments += comments

    # Check if there are more comments to retrieve
    next_page_token = response.json().get("nextPageToken")

# Print the comments
# result = [comment["snippet"]["topLevelComment"]["snippet"]["textDisplay"] for comment in all_comments]
for comment in all_comments:
    print(comment["snippet"]["topLevelComment"]["snippet"]["textDisplay"])
```
